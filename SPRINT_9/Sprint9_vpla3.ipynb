    * Escribe una función que calcule el IMC ingresado por el usuario, es decir, quien lo ejecute tendrá que ingresar estos datos.
    * La función debe clasificar el resultado en sus respectivas categorías.
    * **Consejo**: Intenta validar los datos previamente, para que envíe un mensaje de advertencia si los datos introducidos por el usuario no están en el formato adequado o no toma valores razonables.

```python
# Se indica las importaciones de librerías usadas en los diferentes ejercicios.
import re
from collections import Counter
from pprint import pprint
import numpy as np
import string
# pip install pyperclip
import pyperclip
from collections import defaultdict
```

```python
def pedir_peso():
    while True:
        try:
            peso = float(input('Escribe tu peso en kg.:'))
            if 20 <= peso <= 300:
                return peso
            else: 
                print('El peso debe estar entre 20 kg. y 300 kg.')
        except ValueError:
            print('Por favor ingresar un número válido. (Ejem: 50 o 65.5)')

def pedir_altura():
    while True:
        try:
            altura = float(input('Escribe tu Altura en mts. (Ejem: 1.65)'))    
            if 0.5 <= altura <= 2.5:
                return altura
            else:
                print('La altura debe estar entre 0.5 mts. y 2.5 mts.')
        except ValueError:
            print('Por favor ingresar un número válido. (Ejem: 1.65)')         

def calcular_imc(peso, altura):
    return round(peso / (altura ** 2), 2)

def clasificar_imc(imc):
    if imc < 18.5: 
        return 'Bajo peso'
    elif imc <= 24.9:
        return 'Peso normal'
    elif imc <= 29.9:
        return 'Sobrepeso'
    else:
        return 'Obesidad'

def calculadora_imc():
    peso = pedir_peso()
    altura = pedir_altura()
    imc = calcular_imc(peso,altura)
    categoria = clasificar_imc(imc)
    print(f'Tu IMC es: {imc}, - {categoria}')
    return imc

calculadora_imc()
```

2. Convertidor de temperaturas
    * Existen diversas unidades de temperatura utilizadas en diferentes contextos y regiones. Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). También existen otras unidades como Rankine (°Ra) y Réaumur (°Re).
    * Selecciona almenos 2 conversiones, de tal manera que al introducir una temperatura retorne, como mínimo, dos conversiones, de manera que se puedan guardar (recuerda que un print() no se puede guardar nunca).
    * **Consejo**: Intenta validar los datos previamente, para que envíe un mensaje de advertencia si los datos introducidos por el usuario no están en el formato adequado.
    * **(EXTRA)**: Piensa una manera de almacenar todas las posibles conversiones en un solo objeto (Lista? Diccionario? DataFrame?) en lugar de escribir muchas condiciones if else en función de la temperatura de origen y la temperatura de destino.

```python
def a_celsius(valor,unidad):    
    match unidad:
        case 'c':
            celsius = valor 
        case 'f':
            celsius = (valor - 32) * 5/9 
        case 'k':
            celsius = valor - 273.15
        case 'ra':
            celsius = (valor * 5/9) - 273.15 
        case 're':
            celsius = valor * 5/4
    return round(celsius, 1)

def desde_celsius(celsius):
    return {
        'C'  : celsius,
        'F'  : round(32 + (celsius * 9/5), 1),  
        'K'  : round(celsius + 273.15, 1),
        'Ra' : round((celsius + 273.15) * 9/5, 1),
        'Re' : round(celsius * 4/5, 1),
    }

def convertir(valor, unidad):
    celsius = a_celsius(valor, unidad)
    return desde_celsius(celsius)

def entrada(texto):
    try:
        numero, unidad = re.split(r"(?=[A-Za-z])", texto, maxsplit=1)
    except ValueError:
        raise ValueError('Formato no válido. Introducir por ej. 27 C, 40 F')
    
    numero = numero.replace(",", ".").strip()
    unidad = unidad.strip().lower()

    try:
        valor = float(numero)
    except ValueError:
        raise ValueError('Por favor ingrese un número válido')         
               
    if unidad not in ("c", "f", "k", "ra", "re"):
        raise ValueError('Unidad no válida. Intorducir una unidad de temperatura válida:: C, F, K, Ra, Re')
    
    return valor, unidad

def pedir_temperatura_origen():
    while True:
        t = input('Temperatura origen: ') 
        print('Temperatura origen: ', t)

        try:
            valor, unidad = entrada(t)
            return valor, unidad
        except ValueError as e:
            print(e)
    
def main():
    print('Este es un convertidor de temperaturas')
    print('Escribe la temperatura que desees convertir y después escribe la unidad, por ejem. 27 C, o 45 F') 
    print('Las unidades de temperatura que reconozco son: C (Celsius), F (Fahrenheit), K (Kelvin), Ra (Rankine), Re (Reaumur)')
    print('')

    valor, unidad = pedir_temperatura_origen()
    temperaturas = convertir(valor, unidad)

    print('Las conversiones de temperaturas son: ')
    print(temperaturas)

if __name__ == '__main__':
    main()
```

3. Contador de palabras de un texto.
    * Escribe una función que dado un texto, muestre cuantas veces aparece cada palabra. Intenta que se gestionen todas las situaciones posibles que puedan hacer que el programa no funcione correctamente.
    * **(EXTRA)**: Cual es la longitud media de las palabras del texto que has escrito? “Hola ¿cómo va?” debería de retornar (4+4+2) / 3 = 3.3.

```python
def limpiar(texto):
    return re.sub(r'[^a-záéíóúüñ\s]', ' ', texto.lower()).split()

def contador(texto):
    palabras = limpiar(texto)
    return Counter(palabras)

def longitud_media(texto):
    palabras = limpiar(texto)
    if not palabras:
        return 0.0
    
    total_letras = sum(len(i) for i in palabras)
    return total_letras / len(palabras)

def main():
    frase = input('Introduce una frase:')
    totcuenta = contador(frase)
    print("Frecuencia de palabras:", totcuenta)
    longmedia = longitud_media(frase)
    print('La longitud media de palabras del texto es: ', longmedia)

if __name__ == '__main__':
    main()
```

4. Diccionario inverso (con posibilidad de duplicados)
    * Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita al revés, es decir, intercambiando las claves y los valores. Los valores y claves en el diccionario original son únicos; si no es así, la función debería imprimir un mensaje de advertencia, juntamente con una lista con los valores asociados a la clave repetida.

```python
dict_encuesta2 = {'a':1, 'b':3, 'c':9, 'd':7, 'e':9}

def diccio_inverso(diccionario):
    dic_inverso = {}
    for clave, valor in diccionario.items():
        if valor not in dic_inverso:
            dic_inverso[valor] = clave
        else:
            if isinstance(dic_inverso[valor], list):
                dic_inverso[valor].append(clave)   
            else:
                dic_inverso[valor] = [dic_inverso[valor], clave]
    return dic_inverso
       
def obten_duplicados(dic_inverso):
    duplicados = {}
    for valor, claves in dic_inverso.items():
        if isinstance(claves, list) and len(claves) > 1:
            duplicados[valor] = claves
    return duplicados

def aviso_duplicado(duplicados):
    if not duplicados:
        return
    for valor, claves in duplicados.items():
            print(f"Advertencia: hay valores duplicados en el diccionario inverso: Valor '{valor}' está asociado a las claves: {claves}")    

def main():
    dinverso = diccio_inverso(dict_encuesta2)
    duplicados = obten_duplicados(dinverso)
    aviso_duplicado(duplicados)
    print('El contenido del diccionario inverso es: ', dinverso)

if __name__ == '__main__':
    main()
```

1. Contador y organizador de palabras de un texto
    * El cliente quedó contento con el contador de palabras, pero ahora quiere leer archivos TXT y que calcule la frecuencia de cada palabra ordenadas dentro de las entradas habituales del diccionario según la letra con la que comienzan, es decir, las claves deben ir de la A a la Z y dentro de la A debemos ir de la A a la Z. Por ejemplo, para el archivo "tu_me_quieres_blanca.txt" la salida esperada sería: {'a': {'a':3, 'agua':1, 'al':2, 'alba':4}, 'b':{'baco':1, 'bebe':1, 'boca':1},'c': {'carnes':2, 'conservas':1},...}

```python
def leer_archivo(ruta_archivo):
   with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
        return archivo.read()

def limpiar(texto):
    return re.sub(r'[^a-záéíóúüñ\s]', ' ', texto.lower()).split()

def contador(lista_palabras):
    return Counter(lista_palabras)

def agrupa_inicial(repeticiones_palabras):
    dic_inicial = {}
    for i, frec in repeticiones_palabras.items():    
        letra = i[0]
        if letra not in dic_inicial:
            dic_inicial[letra] = {}
        dic_inicial[letra][i] = frec
    return dic_inicial

def ordena_diccionario(dic_inicial):
    dic_final = {}
    for letra in sorted(dic_inicial.keys()):
        dic_final[letra] = {
            i: dic_inicial[letra][i]
            for i in sorted(dic_inicial[letra].keys())
        }
    return dic_final

def conta_palabras_texto(ruta_archivo):
    contenido = leer_archivo(ruta_archivo)
    palabras = limpiar(contenido)
    frecuencia = contador(palabras)
    dic_inicial = agrupa_inicial(frecuencia)
    dic_final = ordena_diccionario(dic_inicial)
    return dic_final

def main():
    ruta = r"C:\Users\vanem\OneDrive\Escritorio\IT Academy - Especializacion\Sprint9\tu_me_quieres_blanca.txt"
    diccionario_texto = conta_palabras_texto(ruta)
    print('El contenido del diccionario es el siguiente: ')
    pprint(diccionario_texto)

if __name__ == '__main__':
    main()
```

1. Generador de contraseñas
    * Explora el funcionamiento del módulo random de la biblioteca numpy.
    * En este punto, el cliente ha detectado un problema con las contraseñas que utilizan sus empleados. Asdf1234, fechas de cumpleaños o similares. Para solucionarlo nos ha encargado una función de Python que genere contraseñas más seguras. La función debe depender de los siguientes parámetros:
        * longitud (int): Longitud de la contraseña
        * mayúsculas (bool = True): Si deben aparecer mayúsculas
        * minúsculas (bool = True): Si deben aparecer minúsculas
        * números (bool = True): Si deben aparecer números
        * signos (bool = False): Si deben aparecer caracteres especiales (,-$? o similares)
    * Así pues, si ejecutamos la función de la siguiente manera:
        * crear_contraseña(10, True, True, True, True)
    * Deberíamos obtener un output (que debemos poder guardar) del estilo:
        * 9Er,5Vn8P$
    * Asegúrate de que se cumplan todos los criterios y de que estas contraseñas sean realmente aleatorias.

    * **(EXTRA)** Explora cómo podríamos hacer que la función copiara la contraseña automáticamente al portapapeles del ordenador (como si la hubiéramos seleccionado y hecho ctrl+copy).

```python
def grupos_caracteres(mayusculas, minusculas, numeros, signos):
    c_mayus = list(string.ascii_uppercase)
    c_minus = list(string.ascii_lowercase)
    c_nums = list(string.digits)
    c_signos = list("!@#$%^&*,-.?")

    alfabeto = []
    grupos_obligatorios = []

    if mayusculas:
        alfabeto.extend(c_mayus)
        grupos_obligatorios.append(c_mayus)
    if minusculas:
        alfabeto.extend(c_minus)
        grupos_obligatorios.append(c_minus)
    if numeros:
        alfabeto.extend(c_nums)
        grupos_obligatorios.append(c_nums)
    if signos:
        alfabeto.extend(c_signos)
        grupos_obligatorios.append(c_signos)
    
    return alfabeto, grupos_obligatorios

def valida_parametros(longitud, grupos_obligatorios):
    if not grupos_obligatorios:
        raise ValueError('Debe activar al menos un tipo de carácter.')
    
    if longitud < len(grupos_obligatorios):
        raise ValueError('La longitud es muy corta para incluir todos los tipos de carácter requeridos.')
    
def elige_uno_por_grupo(grupos_obligatorios):
    contraseña_c = []
    for grupo in grupos_obligatorios:
        idx = np.random.randint(0, len(grupo))
        contraseña_c.append(grupo[idx])
    return contraseña_c

def rellenar_con_alfabeto(contraseña_c, alfabeto, longitud_tot):
    resto = longitud_tot - len(contraseña_c)
    alfabeto_array = np.array(alfabeto)
    indices =np.random.randint(0, len(alfabeto_array), size=resto)
    contraseña_c.extend(list(alfabeto_array[indices]))
    return contraseña_c

def barajar_une(caracteres):
    caracteres_array = np.array(caracteres)
    np.random.shuffle(caracteres_array)
    return "".join(caracteres_array.tolist())

def copia_portapapeles(texto):
    pyperclip.copy(texto)

def crear_contraseña(longitud: int, mayusculas: bool = True, minusculas: bool = True,
                     numeros: bool = True, signos: bool = False):
    alfabeto, grupos_obligatorios = grupos_caracteres(mayusculas, minusculas, numeros, signos)
    valida_parametros(longitud, grupos_obligatorios)
    contraseña_inici = elige_uno_por_grupo(grupos_obligatorios)
    contraseña_relle = rellenar_con_alfabeto(contraseña_inici, alfabeto, longitud)
    contraseña = barajar_une(contraseña_relle)

    copia_portapapeles(contraseña)
    return contraseña

def main():
    gen_c = crear_contraseña(10, True,True, True, True)
    print('Contraseña generada: ', gen_c)
    print('Contraseña copiada al portapapeles, puedes pegarla con Ctrl+V')

if __name__ == '__main__':
    main()
```

2. Procesamiento de datos simple
    * Una compañera de trabajo nos ha pedido un favor, aprovechando que sabe que estamos aprendiendo a programar. Tiene un historial de partidos de fútbol catalán en un archivo, donde se almacenan los nombres de los equipos y los resultados. Necesita que procesemos los datos de manera automática, para extraer los resultados que necesita.
    * Utiliza el archivo "historic_partits.txt"
    * Necesita un programa que devuelva:
        * El número total de goles que ha hecho cada equipo.
        * El nombre del equipo más goleador.
        * El nombre del equipo más goleado.
        * La clasificación general (cada victoria: 3 pts, empate 1 pt, derrota 0 pts)
    * Como referencia, los equipos con más puntos serían los siguientes: (imagen)
    * Puedes utilizar el formato que te sea más conveniente.

```python
def crear_registro():
    return {"w": 0, "d": 0, "l": 0, "gf": 0, "ga": 0, "pts": 0}

def leer_partidos(ruta_archivo):
    with open(ruta_archivo, "r", encoding="utf-8") as f:
        for linea in f:
            linea = linea.strip()
            if not linea:
                continue

            partes = linea.split("\t")
            if len(partes) != 3:
                continue

            team1 = partes[0].strip()
            marcador = partes[1].strip()
            team2 = partes[2].strip()

            try:
                g1_str, g2_str = marcador.split("-")
                g1 = int(g1_str)
                g2 = int(g2_str)
            except ValueError:
                continue

            yield team1, g1, team2, g2

def actualizar_stats(stats, team1, g1, team2, g2):
    stats[team1]["gf"] += g1
    stats[team1]["ga"] += g2
    stats[team2]["gf"] += g2
    stats[team2]["ga"] += g1

    if g1 > g2:
        stats[team1]["w"] += 1
        stats[team2]["l"] += 1
        stats[team1]["pts"] += 3
    elif g2 > g1:
        stats[team2]["w"] += 1
        stats[team1]["l"] += 1
        stats[team2]["pts"] += 3
    else:
        stats[team1]["d"] += 1
        stats[team2]["d"] += 1
        stats[team1]["pts"] += 1
        stats[team2]["pts"] += 1

def resumen_liga(stats):
    if not stats:
        raise ValueError("No se ha leído ningún partido. Revisa el archivo o el formato.")

    equipo_mas_goleador = max(stats, key=lambda t: stats[t]["gf"])
    equipo_mas_goleado = max(stats, key=lambda t: stats[t]["ga"])

    clasificacion = sorted(
        stats.items(),
        key=lambda item: (item[1]["pts"], item[1]["gf"]),
        reverse=True,
    )
    return equipo_mas_goleador, equipo_mas_goleado, clasificacion

def procesar_partidos(ruta_archivo):
    stats = defaultdict(crear_registro)

    for team1, g1, team2, g2 in leer_partidos(ruta_archivo):
        actualizar_stats(stats, team1, g1, team2, g2)

    mas_goleador, mas_goleado, clasificacion = resumen_liga(stats)
    return stats, mas_goleador, mas_goleado, clasificacion

def imprimir_clasificacion(clasificacion):
    print(f"{'Equipo':20} {'ganados':9} {'empatados':9} {'perdidos':9} {'goles':5} {'encajados':10} {'puntos':5}")
    for equipo, s in clasificacion:
        print(f"{equipo:20} {s['w']:9} {s['d']:9} {s['l']:9} {s['gf']:5} {s['ga']:10} {s['pts']:5}")

def main():
    ruta = r"C:\Users\vanem\OneDrive\Escritorio\IT Academy - Especializacion\Sprint9\historic_partits.txt"
    stats, mas_goleador, mas_goleado, clasificacion = procesar_partidos(ruta)

    print("Equipo más goleador:", mas_goleador)
    print("Equipo más goleado:", mas_goleado)
    print()
    imprimir_clasificacion(clasificacion)

if __name__ == "__main__":
    main()
```

    4. Fes una petició POST per crear una nova publicació fictícia. Inclou el títol, el cos del missatge i un userId.

    5. Fes una petició PATCH per modificar parcialment una publicació existent.

```python
actualizar_public = {'title': 'Titulo modificado'}

url_actualiza = f'{url_inicial}/posts/1'
respuesta_patch = requests.patch(url_actualiza, json = actualizar_public)

print('Actualizar post con PATCH:', 'Codigo de estado:', respuesta_patch.status_code)
print('Respuesta JSON:', respuesta_patch.json())
```

# Nivell 3

    API d'Open Data Barcelona
    1. Utilitza l’API d’Open Data BCN per:
        Cercar un dataset del teu interès mitjançant package_search i package_show.

```python
url_base = "https://opendata-ajuntament.barcelona.cat/data/api/3/action"

# Consulta usando package_search - se buscan varios datasets con la palabra "arbra" y un máximo de 10 datasets.
paramet_search = {"q": "arbrat", "rows": 10}

r_search_bcn = requests.get(f"{url_base}/package_search", params = paramet_search)

print("codigo de estado:" , r_search_bcn.status_code)

data_search_bcn = r_search_bcn.json()
```

```python
# Reviso los resultados para elgir un dataset
datasets = data_search_bcn["result"]["results"]

print("Num datasets:", len(datasets))

for ds in datasets:
    print(f"ID: {ds['id']}  |"
        f"Nombre: {ds['name']}  |"
        f"Titulo: {ds['title']}")
```

```python
# Dataset elegido
dataset_id = "9b525e1d-13b8-48f1-abf6-f5cd03baa1dd"

# Utilizo package_show - para ver los detalles de mi dataset
r_show_bcn = requests.get(f"{url_base}/package_show", params = {"id": dataset_id})

print("codigo de estado (package_show):" , r_show_bcn.status_code)

data_show_bcn = r_show_bcn.json()

dataset = data_show_bcn["result"]
```

    2. Dels resultats obtinguts, selecciona’n un que tingui recursos disponibles en CSV o JSON.

```python
# transformo a Dataframe con pandas, para visualizar mejor mi resultado y filtrado.

resources = dataset["resources"]

resumen_resources = [{"id": r["id"], "name": r.get("name"), "format": r.get("format"), "url": r.get("url"),}
    for r in resources]

df_resources =pd.DataFrame(resumen_resources)

df_resources_filtrado = df_resources[df_resources["format"].str.upper().isin(["CSV","JSON"])]
df_resources_filtrado
```

    3. Amb el resource_id del recurs seleccionat, realitza una consulta mitjançant datastore_search per:
        Recuperar almenys 100 registres.

```python
# Recurso selecionado
resource_id = "29cd5c1f-11b1-404b-b3a5-ae29940b8c55"

# Consulta con datastore_search
params_data = {"resource_id": resource_id, "limit": 100}

r_data = requests.get(f"{url_base}/datastore_search", params = params_data)

print("codigo de estado (datastore_search):", r_data.status_code)

data = r_data.json()

records = data["result"]["records"]
print("Registros recuperados:", len(records))
```

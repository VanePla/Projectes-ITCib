se descargan los archivos .7zip y se descomprimen para poder leer los .csv

```python
def _nombre_7z_desde_url(url: str) -> str:
    """Obtiene un nombre único .7z desde la URL del recurso."""
    return url.split("/")[-2] + ".7z"

def descargar_7z(url: str, carpeta: Path) -> Path:
    """Descarga un .7z si no existe y devuelve la ruta local."""
    nombre = _nombre_7z_desde_url(url)
    ruta   = carpeta / nombre
    if not ruta.exists():
        resp = requests.get(url, timeout=60)
        resp.raise_for_status()
        ruta.write_bytes(resp.content)
    return ruta

def extraer_7z(ruta_7z: Path, carpeta_destino: Path) -> Path:
    """Descomprime un .7z en carpeta propia (si aún no está) y devuelve la carpeta."""
    carpeta = carpeta_destino / ruta_7z.stem
    carpeta.mkdir(exist_ok=True)

    # Solo extrae si la carpeta está vacía
    if not any(carpeta.iterdir()):
        with py7zr.SevenZipFile(ruta_7z, mode="r") as z:
            z.extractall(path=carpeta)
    return carpeta

def descargar_y_extraer_bicing(url: str, carpeta: Path) -> Path:
    """Descarga un .7z de Bicing y lo descomprime, devuelve carpeta con CSVs."""
    ruta_7z   = descargar_7z(url, carpeta)
    carpeta_d = extraer_7z(ruta_7z, carpeta)
    return carpeta_d

def descargar_todos_bicing(urls_por_anio: dict, carpeta: Path) -> None:
    """Descarga y descomprime todos los .7z de todos los años."""
    for anio, urls in sorted(urls_por_anio.items()):
        print(f"\n📥 Año {anio}")
        for i, url in enumerate(urls, 1):
            carpeta_mes = descargar_y_extraer_bicing(url, carpeta)
            print(f"  [{i:02d}/{len(urls)}] {carpeta_mes.name}")

# Ejecutar solo una vez
descargar_todos_bicing(bicing_urls, Data_bicing_raw)
```

se crea dfgeo de solo distritos

```python
# filtro filas de tipo distrito
gdf_dist = gdf_admin[(gdf_admin["TIPUS_UA"] == "DISTRICTE")].copy()

# renombro columnas
gdf_dist = gdf_dist.rename(columns={"CODI_UA": "codigo_distrito", "NOM": "nombre_distrito"})
gdf_dist = gdf_dist[["codigo_distrito", "nombre_distrito", "geometry"]]

gdf_dist =gdf_dist.to_crs("EPSG:4326")

gdf_dist.to_csv("data_procesada/distritosBCN.csv", index=False)
gdf_dist.head()
```

```python
info_geo = info_bicing_estac_peq.copy()

gdf_geo = gpd.GeoDataFrame(info_geo, geometry=gpd.points_from_xy(info_geo["lon"], info_geo["lat"], crs="EPSG:4326"))

gdf_geo.head()
```

se hace Join espacial estaciones-distritos

```python
gdf_espa_dist = gpd.sjoin(gdf_geo, gdf_dist, how="left", predicate="within")
gdf_espa_dist[["id_estacion", "nombre_calle", "lat", "lon", "codigo_distrito", "nombre_distrito"]].head()
```
